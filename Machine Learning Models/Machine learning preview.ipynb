```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df= pd.read_csv('df_ETL.csv')
```

The first thing is to return the datasets to the way they were before cleaning to be able to work on them separately.

```python
df.shape
```

```python
train =df[:1366]
test = df[1366:]
```


Now we will split our training dataset again to see how the models behave with themselves and then we will evaluate them against the test dataset.

```python
y = train['SalePrice']
X = train.drop(columns = 'SalePrice')
```

```python
#Divining the datasets into training and testing
X_train, X_test, y_train, y_test = train_test_split(X,y,)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_prediction = model.predict(X_test)
y_prediction
```

```python
mean_squared_error(y_test, y_prediction)
```

```python
r2_score(y_test, y_prediction)
```
